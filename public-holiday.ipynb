API Data

```python
import requests
import pandas as pd
import datetime

baseurl = 'https://date.nager.at/api/v3/publicholidays/2023/AT'

def fetch_data_from_api(baseurl):
    try:
        response = requests.get(baseurl)
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Failed to fetch data from API. Status Code: {response.status_code}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"An error occurred while fetching data: {e}")
        return None

# fetch data from the API
data_from_api = fetch_data_from_api(baseurl)

if data_from_api:
    # create dataFrame using the fetched data
    df = pd.DataFrame(data_from_api)
    
    # day (Monday, Tuesday, Friday etc.) each of those holiday falls in
    df['date'] = pd.to_datetime(df['date'])
    df['day_of_week'] = df['date'].dt.day_name()

else:
    print("Data not available.")

display(df)
```
